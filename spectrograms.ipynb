```python
dataset = DatasetHARUML20(mode = 'leave-one-subject')
dataset.loadData()
```

```python
import matplotlib.pyplot as plt

x = dataset.X_train[0, :, 0]

# print(signal)
plt.plot(x)
```

```python
from scipy.fft import fftshift
import scipy
from scipy import  signal
```

```python


f, t, Sxx = signal.spectrogram(x, 50, nperseg = 50)
# f, t, Sxx = signal.spectrogram(x, 50, nperseg = 20)

plt.figure(figsize=(8,10))
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
import librosa
```

```python
S = librosa.feature.melspectrogram(y=x, sr=1/50, n_fft=180)
```

```python
import librosa.display

fig, ax = plt.subplots()

S_dB = librosa.power_to_db(S, ref=np.max)

img = librosa.display.specshow(S_dB, x_axis='time',

                         y_axis='mel', sr=1/50,

                         fmax=8000, ax=ax)

fig.colorbar(img, ax=ax, format='%+2.0f dB')

ax.set(title='Mel-frequency spectrogram')
```
